```python
def NeufeldSolution(x,x_input=0.0,a=1.0,tau=1.0):
    intense  = x**2 * (sqrt(6.0)/24.0) * (1.0/(a*tau))
    arg = abs(x**3 - x_input**3)
    arg = sqrt(pi**4 / 54.0) * arg / (a * tau)
    intense = intense / cosh(arg)
    norma = trapz(intense, x)
    intense  = intense/norma # the integral is normalized to be one
    return intense 
```

# The interpolation setup
I will define two optical depths $\tau_a=10^6$ and $\tau_b=10^7$. I will interpolate the solution for $\tau = 6\times10^6$.

I calculate the analytical solutions for the three. Then I construct a Monte Carlo realization for the $\tau_a$ and $\tau_b$.

From these two MonteCarlo realization I interpolate the corresponding spectra for $\tau_c$.

```python
delta_x = 0.01
tau_a = 1.0E6
x=arange(-50.0,50.0,delta_x)
J_a=NeufeldSolution(x,a=1.0e-3,tau=tau_a)
plot(x,J_a)
```

```python
def SampleNeufeldSolution(x_input=0.0,a=1.0,tau=1.0, n_initial=1000000, n_final=100000):
    #generates random frequencies
    min_x = -45.0 + x_input
    max_x = 45.0 + x_input
    x_in = random.random(n_initial)
    x_in = (x_in*(max_x-min_x)) + min_x
    
    #generate the intensity for the Neufeld solution
    neufeld = NeufeldSolution(x_in,x_input=x_input,tau=tau,a=a)
    
    #generate random points in the y-direction
    min_y = 0.0
    max_y = max(neufeld)*1.1
    y = random.random(n_initial)
    y = (y*(max_y-min_y)) + min_y

    #select the points under the cruve
    below = where(y<neufeld)
    print below, size(below)
    below = below[0]
    cut_below = below[0:n_final]
    x_mc = x_in[cut_below]
    x_ms = sort(x_mc)
    return x_ms
```

#Interpolation
In the interpolation we assign to a new MonteCarlo array new values of $x_{\rm int}$ from the $x_a$ and $x_{b}$ that have the same value of $F(>x)$.

The key of a good interpolation is the right estimate of $\mathtt{weight}$. Here I assume that it is linear in $\log_{10}\tau$

```python
weight = (log10(tau_c) - log10(tau_a))/(log10(tau_b) - log10(tau_a))
print weight
x_sample_c = x_sample_a  + (weight * (x_sample_b - x_sample_a))
integral_c = linspace(1.0,0.0,x_sample_c.size)
```
