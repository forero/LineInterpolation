```python
def NeufeldSolution(x,x_input=0.0,a=1.0,tau=1.0):
    intense  = x**2 * (sqrt(6.0)/24.0) * (1.0/(a*tau))
    arg = abs(x**3 - x_input**3)
    arg = sqrt(pi**4 / 54.0) * arg / (a * tau)
    intense = intense / cosh(arg)
    return intense
```

```python
delta_x = 0.01
tau_a = 1.0E6
x=arange(-50.0,50.0,delta_x)
J_a=NeufeldSolution(x,a=1.0e-3,tau=tau_a)
plot(x,J_a)
```
